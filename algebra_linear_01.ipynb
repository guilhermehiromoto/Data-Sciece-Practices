## Vetores

### O que são vetores?

Em [_"Essence of Linear Algebra"_](http://3b1b.co/eola) é levantado 3 abordagens do que é um vetor:

1. Na física, um vetor é uma "seta" de determinado comprimento (magnitude), direção e sentido.
2. Na ciência da computação, um vetor é uma lista ordenada de números.
3. Na matemática, vetor são objetos num certo espaço com dadas operações de soma e produto por escalar: $\mathbf{u} + \mathbf{v}, \alpha\cdot\mathbf{v}$.

<img src="./vector.jpg" style="width: 500px;"/>

Na ciência da computação e na ciência de dados, os vetores costumam ser multidimensionais, ou seja, têm muitos componentes. 

Um vetor pode, por exemplo, representar os dados de uma pessoa.

### Operações fundamentais de vetores

### Soma de vetores

Suponha dois vetores: 

$$
   \mathbf{a} = \left[ \begin{array}{c} -2 \\ 1  \end{array} \right], \quad  
   \mathbf{b} = \left[ \begin{array}{c} 1 \\ -3  \end{array} \right] 
$$

```python
a = np.array((-2,1))
b = np.array((1,-3))
origin = np.array((0,0))

vectors = [a, b, a+b]
tails   = [origin, a, origin]
plot_vector(vectors, tails)
```

$$
   \left[ \begin{array}{c} -2 \\ 1  \end{array} \right] +  
   \left[ \begin{array}{c} 1 \\ -3  \end{array} \right] =
   \left[ \begin{array}{c} -2+1 \\ 1-3  \end{array} \right]
$$


### Multiplicação por escalar

Seja o escalar $2$ e o vetor $\mathbf{c} = \left[ \begin{array}{c} 2 \\ 1  \end{array} \right]$ :

```python
c = np.array((2,1))
vectors = [c, 2*c]
plot_vector(vectors)
```

$$
  2\cdot\left[ \begin{array}{c} 2 \\ 1  \end{array} \right] =
  \left[ \begin{array}{c} 2\cdot 2 \\ 2\cdot 1  \end{array} \right] = 
  \left[ \begin{array}{c} 4 \\ 2  \end{array} \right]
$$

## Vetores de base e Coordenadas

Dado um vetor $(x,y)$ qualquer, temos que:

<br>

$$
 \left[ \begin{array}{c} x \\ y  \end{array} \right] =
 x\cdot\left[ \begin{array}{c} 1 \\ 0  \end{array} \right] +
 y\cdot\left[ \begin{array}{c} 0 \\ 1  \end{array} \right]
$$

<br>

O conjunto formado pelos vetores $ \,\mathbf{\hat{i}}=\left[ \begin{array}{c} 1 \\ 0  \end{array} \right]$ e $\,\mathbf{\hat{j}}=\left[ \begin{array}{c} 0 \\ 1  \end{array} \right] $ é um exemplo de base. Os escalares x e y acima são o que chamamos de coordenadas do vetor nessa base.

<br>

Para ser uma base, os vetores que a compõe precisam:
 - Ser linearmente independentes
 - Gerar qualquer outro vetor do espaço

```python
i = np.array((1,0))
j = np.array((0,1))

vec = 3*i + 2*j
vectors = [i, j, 3*i, 2*j, vec]
plot_vector(vectors)
```

### Combinação linear e Span

Dado dois vetores $\vec{u}$ e $\vec{v}$, toda combinação do tipo $\,\,\alpha\cdot \vec{u} + \beta\cdot \vec{v}\,\,$ é denominada **combinação linear**.

Todo vetor é uma combinação linear dos vetores de uma determinada base.

Um **span** representa todas as combinações lineares possíveis de um conjunto de vetores.

```python
vectors = []
i = np.array((1,0))
j = np.array((0,1))

for _ in range(30):
    m = randint(-10,10)
    n = randint(-10,10)
    vectors.append(m*i + n*j)
    
plot_vector(vectors)
```

Gerando o span de $\mathbf{a}$ e $\mathbf{b}$ definidos anteriormente:

```python
vectors = []
for _ in range(30):
    m = randint(-10,10)
    n = randint(-10,10)
    vectors.append(m*a + n*b)
    
plot_vector(vectors)
```

Agora com $\mathbf{a}$ e $\mathbf{d} = \left[ \begin{array}{c} -1 \\ 0.5  \end{array} \right]$

```python
d = np.array((-1,0.5))
vectors = []
for _ in range(30):
    m = randint(-10,10)
    n = randint(-10,10)
    vectors.append(m*a + n*d)
    
plot_vector(vectors)
```

> No caso de uma **base** de vetores, o seu **span** representa todo o seu espaço.

Afinal, o que é ser linearmente independente?

> Podemos pensar em um conjunto como sendo LI quando nenhum vetor pode ser escrito como uma combinação linear dos demais.

No 3D, o span de dois vetores LI geram um plano. Adicionando mais um terceiro vetor que não é combinação linear dos outros dois, temos que o span deles gera todo o espaço.

## Matrizes

### O que é uma matriz?

É possível entender matrizes de 3 maneiras:
- Tabela de números
- Conjuntos de vetores
- Representação de uma transformação linear


### Matrizes com NumPy

$A$  = $
\begin{bmatrix} -2 & 1 \\ 
                    1 & -3  \end{bmatrix}  
$

```python
A = [[-2,1], [1,-3]]
A = numpy.array(A)
print(A)
```

Multiplicando a matriz $A$ pelo vetor $\mathbf{c} = \left[ \begin{array}{c} 2 \\ 1  \end{array} \right]$:

```python
#vectors = [c]
#plot_vector(vectors)

c_l = A.dot(c)
vectors = [c, c_l]
plot_vector(vectors)
```

$A$  = $
\begin{bmatrix} 0 & 1 \\ 
                    -1 & 0  \end{bmatrix}  
$

```python
#vectors = [c]
#plot_vector(vectors)

A = np.array([[0,1],[-1,0]])

c_l = A.dot(c)
vectors = [c, c_l]
plot_vector(vectors)
```

Em ${\rm I\!R}^{2}$, seja a base canônica $ B =\{(1,0), (0,1)\}$
e o vetor $\mathbf{c} = \left[ \begin{array}{c} 2 \\ 1  \end{array} \right]$

```python
c = np.array([2,1])
vectors = [c]
plot_vector(vectors)
```

Como podemos representar esse vetor em uma nova base $B' = \{ (1,1), (0,1)\}$ ?

```python
A = np.array([[1,0],[-1,1]])
c_l = [A.dot(np.array(c))]

plot_vector(c_l)
```
